```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data_corr = df.select_dtypes("int64","float64").corr()
```

```python
sns.heatmap(data_corr,annot = True)
plt.show()
```

```python
sns.heatmap(data_cov,annot = True)
plt.show()
```
